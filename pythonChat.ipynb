```python
#Datta loading first step
from langchain_community.document_loaders import TextLoader

loader = TextLoader("Introduction_to_Python.txt", encoding="utf-8")
text_data = loader.load()
text_data
```

```python
# Text splitter
from langchain_text_splitters import RecursiveCharacterTextSplitter
textSplitter = RecursiveCharacterTextSplitter(chunk_size=200,chunk_overlap=20)
finalDocuments = textSplitter.split_documents(text_data)
finalDocuments
```

```python
# If finalDocuments is a list of Document objects:
texts = [doc.page_content for doc in finalDocuments]
```

```python
##Huggingface embedding
import os
from dotenv import load_dotenv
load_dotenv()
# Set HF_Token in your .env file instead of hardcoding it here
```

```python
from dotenv import load_dotenv
import os

load_dotenv()
hf_token = os.getenv("HF_Token")

print(hf_token)  # should print your token
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
print(embeddings)
```

```python
# store in vector db
from langchain_community.vectorstores import FAISS
db = FAISS.from_documents(finalDocuments, embeddings)
db
```

```python
query = input("Enter you question for Python : ")
docs = db.similarity_search(query)
for i in range(len(docs)):
    print(docs[i].page_content)
```
