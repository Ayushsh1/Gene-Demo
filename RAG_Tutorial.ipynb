```python
#API based loader
# import requests
# from langchain.schema import Document

# response = requests.get("www.themealdb.com/api/json/v1/1/search.php?f=a")
# documents=[
#     Document(page_content=item["text"])
#     for item in response.json()
# ]
# documents

import requests
from langchain_core.documents import Document

# Always include https://
response = requests.get("https://www.themealdb.com/api/json/v1/1/search.php?f=a")

data = response.json()

# Extract the list of meals
meals = data.get("meals", [])

# Convert each meal into a Document
documents = [
    Document(page_content=meal["strMeal"], metadata=meal)
    for meal in meals
]

print(documents[0])


```

```python
#Fixed-Size Chunking
# Description
# Split text into chunks of fixed length (characters or tokens).
from langchain_text_splitters import CharacterTextSplitter

splitter = CharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)
# If txt_data is a list of strings, join them
text_data = " ".join([doc.page_content for doc in text_data])
chunk = splitter.split_text(text_data)
chunk[0]
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(
    chunk_size=800,
    chunk_overlap=100
)

chunks = splitter.split_documents(document_data)

for i, chunk in enumerate(chunks, start=1):
    print(f"Chunk {i}:\n{chunk.page_content}\n{'-'*40}")
```

```python
# Token-Based Chunking (LLM-Aware)
# Description
# Splits text by token count, not characters.   
from langchain_core.documents import Document
from langchain_text_splitters import TokenTextSplitter

splitt = TokenTextSplitter(chunk_size=250, chunk_overlap=50)

# txt_data is a list of Document objects
chunks = splitt.split_documents(document_data)

for i, chunk in enumerate(chunks, start=1):
    print(f"Chunk {i}:\n{chunk.page_content}\n{'-'*40}")
```
